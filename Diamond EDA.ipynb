#### To sort the values of categoric variables as "ordinal"

```python
# from the worst (fair) to the best (ideal)

cut_categories = ["Fair", "Good", "Very Good", "Premium", "Ideal"]
diamond['cut'] = pd.Categorical(diamond['cut'], categories=cut_categories, ordered=True)
```

```python
cut_categories
```

```python
clarity_categories = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
diamond["clarity"] = pd.Categorical(diamond['clarity'],categories = clarity_categories,ordered=True)
                                              
```

```python
clarity_categories
```

#### Data visualization

##### bar plot

```python
#calculate frequencies
cut_frequencies = diamond["cut"].value_counts(ascending=True)
# Extract x and y values for the plot
x = cut_frequencies.values
y = cut_frequencies.index

# Plot the bar chart
plt.bar(y, x, color='red')
plt.title("Class frequencies of 'cut' variable")
plt.xlabel('Frequency')
plt.ylabel('Cut Category')
plt.show()
```

```python
# For 'cut' variable
sns.barplot(x = diamond["cut"].value_counts().index,
            y = diamond["cut"].value_counts(),
            data = diamond,
            saturation = 1);
```

```python
# For 'color' variable
sns.barplot(x = diamond["color"].value_counts().index,
            y = diamond["color"].value_counts(),
            data = diamond,
            saturation = 1);
plt.title("Barplot for diamond")
```

```python
sns.catplot(x = "cut",
            y = "price",
            kind = "strip",
            data = diamond,
            margin_titles = True);
```

```python
sns.catplot(x = "clarity",
            y = "price",
            kind = "strip",
            data = diamond,
            margin_titles = True);
```

```python
sns.catplot(y="price",
            x="clarity",
            kind="box",
            color= '#abc9ea',
            data=diamond).set(yscale="log");
plt.title("Boxplot of Diamond Prices by Clarity")
plt.show()
```

```python
sns.barplot(x = "cut", y = "price", hue = "color", data = diamond, saturation = 1);
plt.title("Average Price by Cut and Color")
plt.show()
```

```python
# Here we validate what our last graphic show us
diamond.groupby(["cut", "color"])["price"].mean()
```

```python
sns.barplot(x = "cut", y = "price", hue = "clarity", data = diamond,
            saturation = 1);
plt.title("Average Price by Clarity and Color")
plt.show()
```

```python
# Here we validate what our last graphic show us

diamond.groupby(["cut", "clarity"])["price"].mean()
```

### Histogram and density plot

```python
(sns.FacetGrid(diamond,
              hue = "cut",
              height = 7,
              xlim = (0, 10000))
 .map(sns.kdeplot, "price", shade = True)
 .add_legend()
);
```
